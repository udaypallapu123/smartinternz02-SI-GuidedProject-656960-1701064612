```python
y=dataset['price_rupiah']
X=dataset.iloc[:,0:7].values
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,y_train,ytest=train_test_split(X,y,test_size=0.2,random_state=40)
```

```python
X_train.shape
```

```python
x_test.shape
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion="entropy",random_state=40)
dt.fit(X_train,y_train)
dtpred=dt.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
def decision_tree(X_train,x_test,y_train,ytest):
  mse=mean_squared_error(ytest,dtpred)
  mae=mean_absolute_error(ytest,dtpred)
  r2=r2_score(ytest,dtpred)
  rmse=np.sqrt(mse)
  print("Mean squared Error (MSE):",mse)
  print("Mean Absolute Error (MAE):",mae)
  print("R-squared:",r2)
  print("Root Mean Squared Error(RMSE):",rmse)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=10,criterion="entropy",random_state=40)
rf.fit(X_train,y_train)
rfpred=rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
def random_forest(X_test,x_train,y_train,ytest):
  mse=mean_squared_error(ytest,rfpred)
  mae=mean_absolute_error(ytest,rfpred)
  r2=r2_score(ytest,rfpred)
  rmse=np.sqrt(mse)
  print("Mean squared Error (MSE):",mse)
  print("Mean absolute Error (MSE):",mae)
  print("R-squared:",r2)
  print("Root Mean Squared Error(RMSE):",rmse)
```

```python
#xgboost regression
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
X_train,xtest,y_train,ytest=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
def xgboost(X_train,xtest,y_train,ytest):
  xgb_reg = xgb.XGBRegressor()
  xgb_reg.fit(X_train,y_train)
  y_pred=xgb_reg.predict(x_test)
  mse=mean_squared_error(ytest,y_pred)
  mae=mean_absolute_error(ytest,y_pred)
  r2=r2_score(ytest,y_pred)
  rmse=np.sqrt(mse)
  print("Mean squared Error (MSE):",mse)
  print("Mean absolute Error (MSE):",mae)
  print("R-squared:",r2)
  print("Root Mean Squared Error(RMSE):",rmse)
```

```python
#testing
xgb_reg=xgb.XGBRegressor()
xgb_reg.fit(X_train,y_train)
xgb_reg.predict([[3,2,13,4,1,1,0]])
```

```python
def comparison(X_train,x_test,y_train,ytest):
  xgboost(X_train,x_test,y_train,ytest)
  print('-'*100)
  random_forest(X_train,x_test,y_train,ytest)
  print('-'*100)
  decision_tree(X_train,x_test,y_train,ytest)
  print('-'*100)
```
